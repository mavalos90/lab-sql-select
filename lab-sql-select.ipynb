## Challenge 1 - Who have published what at where

Write a SELECT query that joins various tables to figure out what titles each author has published at which publishers.

```python
query= ('''SELECT a.au_id AS AUTHOR_ID, a.au_lname AS LAST_NAME, a.au_fname as FINAL_NAME, t.title AS TITLE, p.pub_name AS PUBLISHER
FROM publications.authors as a
LEFT JOIN publications.titleauthor as ta ON a.au_id = ta.au_id
LEFT JOIN publications.titles as t ON t.title_id = ta.title_id
LEFT JOIN publications.publishers as p ON p.pub_id = t.pub_id
ORDER BY AUTHOR_ID''')
```

```python
query_job = client.query(query=query)
query_job.to_dataframe()
```
